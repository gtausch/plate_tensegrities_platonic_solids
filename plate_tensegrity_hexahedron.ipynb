```python
# Algebraic treatment of a Symmetric Plate-Tensegrity-Structure with the 432 Rotational Spherical Symmetry of a Hexahedron. 
# GT, 2026-05-23

# 1. Description of the Structure
# We start with a cube with side-length 2/sqrt(2) such that its vertices are of distance 1 to the barycenters of adjacent sides.
# The cube as a tensegrity structure made of plate consists of 8 x 3 ring cables and 3 x 4 edge-cables as well as 6 equilateral quadratic plates. 
# At each node exactly one vertex of a plate and three cables meet. Two of the cables at each point are ring cables, one is an edge-cable. 
# By definition all 24 nodes or points are equal up to symmetry.
# We orient the cube in space such that the center of gravity of each sides is on either the  x, y or z-axis.
# The center of gravity of the cube is the origin.

# 2. What we want to derive 
# We are starting with a closed  polyhedron and are looking for 3 motors composed of rotations and expansions that define a tensegrity-equilibrium for any given rotation alpha of the top side. 
# Since opposite sides have opposite rotations and all sides are same up to symmetry, we have only one unknown the expansion given the rotation alpha.
# The rotation angle alpha can be between 0 and 45° degrees and defines a rotation of all plates around the orthomal axis through the center of gravity of the plate.
# Tensegrity equilibria are possible for any alpha between 0 and 45° degrees.
# Globally the tensegric configuratations transform from a cube to a cuboctahedron in an expanding rotation. Plate sizes remain fixed but the cables change lengths.

# 3. Geometric Setup and Orientation of the Fundamental Region 
# Since dealing algebraically with the rotations in 3D is cumbersome in usual matrix formulation, we use geometric algebra.
# In Geometric Algebra P(3,0,1) we can express all rotations as rotors in a very compact form. 

# 4. Programming Setup
# We use the Kingdon package for Python to perform algebraic operations in Geometric Algebra P(3,0,1).
# Sympy is used to perform general algebraic manipulations.
# Math in Python is used to perform numerical calculations for the visualization of specific solutions. 
# The Ipykernel and Ipywidgets and the Ipython.display packages are used for visualization. 
```

```python
# Setting up the programming environment by importing the necessary packages.
from kingdon import Algebra
import sympy as sym
from math import cos, sin, pi, sqrt
import ipywidgets
from IPython.display import display
from ipywidgets import FloatSlider
```

```python
# Setting up the Projective Algebra PGA (3,0,1)
# For representation of points, lines and planes in PGA (3,0,1) please refer to the Geometric Algebra Cheat Sheet: https://bivector.net/3DPGA.pdf
# For the Kingdon-Syntax please refer to the documentation: https://kingdon.readthedocs.io/en/latest/

alg = Algebra(3, 0, 1)
locals().update(alg.blades)

# Points represent nodes 
point = lambda x, y, z: (e0 + x*e1 + y*e2 + z*e3).dual()
# Lines represent cables with zero diameter
line = lambda x1, y1, z1:(e0 + x1*e12 + y1*e13 + z1*e23)
# Planes represent plates with zero thickness
plane = lambda a, b, c, d: alg.vector(e1=a, e2=b, e3=c, e0=d)
```

```python
a = 2/sym.sqrt(2)
b = 2/sym.sqrt(2)
c = 2/sym.sqrt(2)

# Variable expansion
s1 = sym.symbols ('s1', real = True)

# Variable for rotation-angle
alpha = sym.symbols ('alpha', real = True)


```

```python
# Points and Plates at zero roation and expansion representing the closed box
# Top
p11 = point(a/2, b/2, c/2) 
p12 = point(-a/2, b/2, c/2) 
p13 = point(-a/2, -b/2, c/2)
p14 = point(a/2, -b/2, c/2)

# Bottom
p64 = point(a/2, b/2, -c/2) 
p63 = point(-a/2, b/2, -c/2) 
p62 = point(-a/2, -b/2, -c/2)
p61 = point(a/2, -b/2, -c/2)

# Right
p21 = point(a/2, b/2, c/2) 
p22 = point(a/2, b/2, -c/2) 
p23 = point(a/2, -b/2, -c/2)
p24 = point(a/2, -b/2, c/2)

# Left
p34 = point(-a/2, b/2, c/2) 
p33 = point(-a/2, b/2, -c/2) 
p32 = point(-a/2, -b/2, -c/2)
p31 = point(-a/2, -b/2, c/2)

# Front
p41 = point(-a/2, -b/2, c/2) 
p42 = point(a/2, -b/2, c/2) 
p43 = point(a/2, -b/2, -c/2)
p44 = point(-a/2, -b/2, -c/2)

# Back 
p54 = point(-a/2, b/2, c/2) 
p53 = point(a/2, b/2, c/2) 
p52 = point(a/2, b/2, -c/2)
p51 = point(-a/2, b/2, -c/2)

```

```python
# Rotors
# Please note the rotors in PGA (3,0,1) need to defined by the half of the angle of rotation.
# Mapping of axes to blades: e12: z-axis, e13: y-axis, e23: x-axis

# first rotor of z-axis
ra1 = sym.symbols('ra1', real=True)
ra1 = sym.cos(alpha/2) +sym.sin(alpha/2) * e12

ra1n = sym.symbols('ra1n', real=True)
ra1n = sym.cos(-alpha/2) +sym.sin(-alpha/2) * e12


# second rotor of x-axis
ra2 = sym.symbols('ra2', real=True)
ra2 = sym.cos(alpha/2) +sym.sin(alpha/2) * e23

ra2n = sym.symbols('ra2n', real=True)
ra2n = sym.cos(-alpha/2) +sym.sin(-alpha/2) * e23


# third rotor of y-axis 
ra3 = sym.symbols('ra3', real=True)
ra3 = sym.cos(alpha/2) +sym.sin(alpha/2) * e13

ra3n = sym.symbols('ra3n', real=True)
ra3n = sym.cos(-alpha/2) +sym.sin(-alpha/2) * e13

```

```python
# Defining the equilibrium conditions based on the coplanarity conditions 
# First coplanarity condition: we look at p12e of the right plate, the center of the top plate mp1e , the center of ring1 m1e and p14e of the top plate 
eq1 = m1e&mp2e&p21e&p52e
eq1 = eq1.values()[0]*4
eq1 = eq1.xreplace({n: 0 for n in eq1.atoms(sym.Float) if abs(n) < 1e-14})
eq1 = eq1.xreplace({n: round(n, 14) for n in eq1.atoms(sym.Float)})
eq1 = sym.nsimplify(eq1)
eq1
```

```python
s1_solve_eq1 = sym.solve(eq1,s1) 
s1_solve_eq1[1]
```

```python


def points_of_box (point, angle):
   
   s1 = s1_solve_eq1[1].evalf(subs={alpha: angle})
   p11 = point
   alpha1 = angle
   
   # Rotors
   # Please note the rotors in PGA (3,0,1) need to defined by the half of the angle of rotation.
   # Mapping of axes to blades: e12: z-axis, e13: y-axis, e23: x-axis

   # first rotor of z-axis
   ra1 = sym.symbols('ra1', real=True)
   ra1 = sym.cos(alpha1/2) +sym.sin(alpha1/2) * e12

   ra1n = sym.symbols('ra1n', real=True)
   ra1n = sym.cos(-alpha1/2) +sym.sin(-alpha1/2) * e12


   # second rotor of x-axis
   ra2 = sym.symbols('ra2', real=True)
   ra2 = sym.cos(alpha1/2) +sym.sin(alpha1/2) * e23

   ra2n = sym.symbols('ra2n', real=True)
   ra2n = sym.cos(-alpha1/2) +sym.sin(-alpha1/2) * e23


   # third rotor of y-axis 
   ra3 = sym.symbols('ra3', real=True)
   ra3 = sym.cos(alpha1/2) +sym.sin(alpha1/2) * e13

   ra3n = sym.symbols('ra3n', real=True)
   ra3n = sym.cos(-alpha1/2) +sym.sin(-alpha1/2) * e13


   # Applying Rotors and Expansions

   # ra1 and e1
   p11e = ra1 >> p11
   p11e.values()[0]= p11e.values()[0]*s1
   p12e = ra1 >> p12
   p12e.values()[0]= p12e.values()[0]*s1
   p13e = ra1 >> p13
   p13e.values()[0]= p13e.values()[0]*s1
   p14e = ra1 >> p14
   p14e.values()[0]= p14e.values()[0]*s1

   p61e = ra1n >> p61
   p61e.values()[0]= p61e.values()[0]*s1
   p62e = ra1n >> p62
   p62e.values()[0]= p62e.values()[0]*s1
   p63e = ra1n >> p63
   p63e.values()[0]= p63e.values()[0]*s1
   p64e = ra1n >> p64
   p64e.values()[0]= p64e.values()[0]*s1

   # ra2 and e2 
   p21e = ra2 >> p21
   p21e.values()[2]= p21e.values()[2]*s1
   p22e = ra2 >> p22
   p22e.values()[2]= p22e.values()[2]*s1
   p23e = ra2 >> p23
   p23e.values()[2]= p23e.values()[2]*s1
   p24e = ra2 >> p24
   p24e.values()[2]= p24e.values()[2]*s1

   p31e = ra2n >> p31
   p31e.values()[2]= p31e.values()[2]*s1
   p32e = ra2n >> p32
   p32e.values()[2]= p32e.values()[2]*s1
   p33e = ra2n >> p33
   p33e.values()[2]= p33e.values()[2]*s1
   p34e = ra2n >> p34
   p34e.values()[2]= p34e.values()[2]*s1

   # ra3 and e3 
   p41e = ra3 >> p41
   p41e.values()[1]= p41e.values()[1]*s1
   p42e = ra3 >> p42
   p42e.values()[1]= p42e.values()[1]*s1
   p43e = ra3 >> p43
   p43e.values()[1]= p43e.values()[1]*s1
   p44e = ra3 >> p44
   p44e.values()[1]= p44e.values()[1]*s1

   p51e = ra3n >> p51
   p51e.values()[1]= p51e.values()[1]*s1
   p52e = ra3n >> p52
   p52e.values()[1]= p52e.values()[1]*s1
   p53e = ra3n >> p53
   p53e.values()[1]= p53e.values()[1]*s1
   p54e = ra3n >> p54
   p54e.values()[1]= p54e.values()[1]*s1

   # midpoints

   # Midpoints of triangular ring cables
   # Top_right_back
   m1e = (p11e+p21e+p53e).normalized()
   # Top_left_back
   m2e = (p12e+p34e+p54e).normalized()
   # Top_left_front
   m3e = (p13e+p31e+p41e).normalized()
   # Top_right_front
   m4e = (p14e+p24e+p42e).normalized()
   # bottom_right_back
   m5e = (p64e+p22e+p52e).normalized() 
   # bottom_left_back
   m6e = (p63e+p33e+p51e).normalized()
   # bottom_left_front
   m7e = (p62e+p32e+p44e).normalized()
   # bottom_right_front
   m8e = (p61e+p23e+p43e).normalized()

   # Top plate center
   mp1e = (p11e+p12e+p13e+p14e).normalized()
   # Right plate center
   mp2e = (p21e+p22e+p23e+p24e).normalized() 
   # Back plate center
   mp3e = (p54e+p53e+p52e+p51e).normalized()
    

   return p11e.map(float), p12e.map(float), p13e.map(float), p14e.map(float), p21e.map(float), p22e.map(float), p23e.map(float), p24e.map(float), p31e.map(float), p32e.map(float), p33e.map(float), p34e.map(float), p41e.map(float), p42e.map(float), p43e.map(float), p44e.map(float), p51e.map(float), p52e.map(float), p53e.map(float), p54e.map(float), p61e.map(float), p62e.map(float), p63e.map(float), p64e.map(float), m1e.map(float), mp1e.map(float), s1




```

```python
# Now we are ready to visualize our tensegrity hexahedron parametrically with respect to rotation angle alpha
# We can make this visualization interactive by defining a function that takes the rotation angle alpha as an argument and updates the coordinates of the points and the graph function accordingly.

def visualize_transformation(alpha_slider, output_widget=None):    

    # Interactive visualization of the tensegrity transformation as a function of rotation angle alpha. 
    if output_widget is not None:
        output_widget.clear_output(wait=True)
   
    # Generate all points based on p11 and the alpha_slider
    p11e, p12e, p13e, p14e, p21e, p22e, p23e, p24e, p31e, p32e, p33e, p34e, p41e, p42e, p43e, p44e, p51e, p52e, p53e, p54e, p61e, p62e, p63e, p64e, m1e, mp1e, ex1 = points_of_box (p11,alpha_slider)

    # Recalculate expansion rate
   

    # Polygon to show coplanarity constraint of the four points at p11. This is not a physical plate but a visual aid to show the coplanarity constraint that is needed for equilibrium at p11.
    coc_c =[[p11e.map(float), m1e.map(float), mp1e.map(float),p24e.map(float)],[p11e.map(float), m1e.map(float)],[m1e.map(float),mp1e.map(float)],[mp1e.map(float), p24e.map(float)],[p24e.map(float), p11e.map(float)]]
    
    # Plates 
    #Top 
    pl1 = [[p11e, p12e, p13e, p14e], [p11e, p12e],[p12e, p13e], [p13e, p14e], [p14e, p11e]]
    #Right    
    pl2 = [[p21e, p22e, p23e, p24e], [p21e, p22e],[p22e, p23e], [p23e, p24e], [p24e, p21e]]
    #Left
    pl3 = [[p31e, p32e, p33e, p34e], [p31e, p32e],[p32e, p33e], [p33e, p34e], [p34e, p31e]]
    #Front 
    pl4 = [[p41e, p42e, p43e, p44e], [p41e, p42e],[p42e, p43e], [p43e, p44e], [p44e, p41e]]
    #Back
    pl5 = [[p51e, p52e, p53e, p54e], [p51e, p52e],[p52e, p53e], [p53e, p54e], [p54e, p51e]]
    #Bottom
    pl6 = [[p61e, p62e, p63e, p64e], [p61e, p62e],[p62e, p63e], [p63e, p64e], [p64e, p61e]]

    #Cable Rings
    cr1 = [[p21e,p53e],[p53e,p11e],[p11e,p21e]]
    cr2 = [[p54e,p34e],[p34e,p12e],[p12e,p54e]]
    cr3 = [[p31e,p41e],[p41e,p13e],[p13e,p31e]]
    cr4 = [[p42e,p24e],[p24e,p14e],[p14e,p42e]]
    cr5 = [[p23e,p43e],[p43e,p61e],[p61e,p23e]]
    cr6 = [[p44e,p32e],[p32e,p62e],[p62e,p44e]]
    cr7 = [[p33e,p51e],[p51e,p63e],[p63e,p33e]]
    cr8 = [[p52e,p22e],[p22e,p64e],[p64e,p52e]]

    # Edge-cables

    ce1 = [p11e,p24e]
    ce2 = [p12e,p53e]
    ce3 = [p13e,p34e]
    ce4 = [p14e,p41e]
    ce5 = [p21e,p52e]
    ce6 = [p54e,p33e]
    ce7 = [p31e,p44e]
    ce8 = [p42e,p23e]
    ce9 =  [p62e,p43e]
    ce10 = [p32e,p63e]
    ce11 = [p51e,p64e]
    ce12 = [p22e,p61e]

     # Tripod for coordinate system
    x_axis_endpoint = point(1.5,0,0)
    y_axis_endpoint = point(0,1.5,0)
    z_axis_endpoint = point(0,0,1.5)  

    # In order to calculate the forces at p11 we recover x,y,z coordinates for all relevant points
    # We recover coordinates of the points from multivectors in PGA (3,0,1) by taking the dual and looking at the coefficients of e1, e2 and e3.
    def point_to_xyz(p):
        # Reverse the dual operation to get back the vector: e0 + x*e1 + y*e2 + z*e3
        v = p.undual()
        # 3. Extract and normalize the coordinates
        x = v.e1
        y = v.e2
        z = v.e3
        x = round (x,16)
        y = round (y,16)
        z = round (z,16)
        return (x, y, z)           

    # We need the points adjacent to p11 
    p11_xyz = sym.Matrix(point_to_xyz(p11e))
    p53_xyz = sym.Matrix(point_to_xyz(p53e))
    p21_xyz = sym.Matrix(point_to_xyz(p21e))
    p24_xyz = sym.Matrix(point_to_xyz(p24e))
    mp1_xyz = sym.Matrix(point_to_xyz(mp1e.normalized()))
    rm1_xyz = sym.Matrix(point_to_xyz(m1e.normalized()))

    # Now we can calculate the proportion of forces for the equilibrium at p11 for this specific configuration.
    # First we need unit vectors for both ring-cables and edge-cables at p11 for this specific configuration, i.e. directions of the forces at p11.

    ring_vec_1 = (p21_xyz - p11_xyz).normalized()
    ring_vec_2 = (p53_xyz - p11_xyz).normalized()
    ring_vec_12 = (ring_vec_1 + ring_vec_2)
    edge_vec_3 = (p24_xyz - p11_xyz).normalized()
    plate_vec_4 = (p11_xyz-mp1_xyz).normalized()


    # The force equilibrium at p11 has to be a closed polygon of forces in 3D space. So we will have to find a scaling operator for each force that allows the polygon to close.
    # We construct a matrix with three unit vectors v12, v3 and v4 and set the scale of v4 to 1 taking directly the coordinates as input. 
    # This leaves us with two unknowns for the scale of v12 and v3.
    # We can solve for the two unknowns by setting the determinant of the matrix to zero, which is the condition for the three vectors to be coplanar and thus in equilibrium.

    force_matrix = sym.Matrix([ 
        [ring_vec_1[0], ring_vec_2[0], edge_vec_3[0], plate_vec_4[0]],
        [ring_vec_1[1], ring_vec_2[1], edge_vec_3[1], plate_vec_4[1]],
        [ring_vec_1[2], ring_vec_2[2], edge_vec_3[2], plate_vec_4[2]],
    ])
    s1 = sym.symbols('s1', real=True)
    s3 = sym.symbols('s3', real=True)
    scale_vec = sym.Matrix([s1, s1, s3, 1])
    force_matrix_sc = force_matrix * scale_vec

    # Solving the force matrix as a system of equations  
    eq_4 = force_matrix_sc[0]
    eq_5 = force_matrix_sc[1]
    eq_6 = force_matrix_sc[2]
    solutions = sym.solve([eq_4, eq_5], (s1, s3))
   
    # Outputing the solution and one set of four forces that are in equilibrium at p11 for this specific configuration. 
    # Forces can be linearly scaled and maintain equilibrium, i.e. we have a prestressable configuration of the structure.

    for key, value in solutions.items():
        var_name = str(key) + "_val"  # e.g., creates "s1_val"
        globals()[var_name] = float(value) 

    scaling_vector = sym.Matrix([s1_val,s1_val, s3_val, 1])
    scaled_forces = force_matrix * scaling_vector
    for i in range(len(scaled_forces)):
        scaled_forces[i] = round(scaled_forces[i], 13)

    ring_vec_1 = ring_vec_1 * s1_val
    ring_vec_2 = ring_vec_2 * s1_val
    ring_vec_1_length = float (ring_vec_1.norm())
    ring_vec_12 = ring_vec_1+ring_vec_2
    ring_vec_12_length = float (ring_vec_12.norm())
    edge_vec_3 = edge_vec_3 * s3_val
    edge_vec_3_length = float (edge_vec_3.norm())
    plate_vec_4 = plate_vec_4 * 1
    edge_vec_4_length = float (plate_vec_4.norm())
    


    # Now we can draw the graph with force vectors at p11 to visualize the equilibrium of forces at this node for this specific configuration.
    # We need to redefine the graph function to include the force vectors at p11.
    # Points in xyz-coordinates for the force vectors
    p1f1_xyz = p11_xyz + ring_vec_1
    p1f2_xyz = p11_xyz + ring_vec_2
    p1f12_xyz = p11_xyz + ring_vec_12
    p1f3_xyz = p11_xyz + edge_vec_3
    p1f4_xyz = p11_xyz - plate_vec_4

    p1f1 = point(p1f1_xyz[0], p1f1_xyz[1], p1f1_xyz[2])
    p1f2 = point(p1f2_xyz[0], p1f2_xyz[1], p1f2_xyz[2])
    p1f12 = point(p1f12_xyz[0], p1f12_xyz[1], p1f12_xyz[2]) 
    p1f3 = point(p1f3_xyz[0], p1f3_xyz[1], p1f3_xyz[2])
    p1f4 = point(p1f4_xyz[0], p1f4_xyz[1], p1f4_xyz[2])    

    # Creating the text_lables
    def fmt_vec_text(name, vec):
        # Formats the vector entries to 3 decimal places so they are easy to read
        return f"{name}: [{float(vec[0]):.3f}, {float(vec[1]):.3f}, {float(vec[2]):.3f}]"


    def graph_func():
        return [
        "Tensegrity Hexahedron (α = {:.1f}°)".format(alpha_slider * 180 / pi),

            0x666666, f"Expansion factor of plates: " f"{ex1:.4f}",
            0x666666, fmt_vec_text("f1 (tension)", ring_vec_1),
            0x666666, fmt_vec_text("f2 (tension)", ring_vec_2),
            0x666666, fmt_vec_text("f12 (tension)", ring_vec_12),
            0x666666, fmt_vec_text("f3 (tension)", edge_vec_3), 
            0x666666, fmt_vec_text("f4 (compression)", plate_vec_4),
            0x666666, f"f1/f2 length: {ring_vec_1_length:.4f}",
            0x666666, f"f12 length: {ring_vec_12_length:.4f}",
            0x666666, f"f3 length: {edge_vec_3_length:.4f}", 
            0x666666, f"f4 length: {edge_vec_4_length:.4f}", 

            
            # Plates
            '<G opacity="0.7">',
            0xcccccc, pl1[0],
            0xcccccc, pl2[0],  
            0xcccccc, pl3[0],
            0xcccccc, pl4[0],
            0xcccccc, pl5[0],
            0xcccccc, pl6[0],
            '</G>',  

            # Coplanarity constraint polygon
            '<G opacity="0.3">',
            0x888800, coc_c[0],
            '</G>',

            # Axes x, y and z 
            '<G opacity="0.8" stroke-dasharray="0.015,0.03">',
            0x330000, [p00.map(float),x_axis_endpoint.map(float)], "x-axis",
            0x330000, [p00.map(float),y_axis_endpoint.map(float)], "y-axis",
            0x330000, [p00.map(float),z_axis_endpoint.map(float)], "z-axis",  
            '</G>',

            # Polygon showing coplanarity
            0x888800, coc_c[1],
            0x888800, coc_c[2],
            0x888800, coc_c[3],
            0x888800, coc_c[4],
            
            # Points
            0x555555, p11e.map(float), "punkt 1",
            0x555555, p12e.map(float), #"p12",
            0x555555, p13e.map(float), #"p13",
            0x555555, p14e.map(float), #"p14",

            0x555555, p21e.map(float), #"p21",
            0x555555, p22e.map(float), #"p22",
            0x555555, p23e.map(float), #"p23",
            0x555555, p24e.map(float), "punkt_2",

            0x555555, p31e.map(float), #"p31",
            0x555555, p32e.map(float), #"p32",
            0x555555, p33e.map(float), #"p33",
            0x555555, p34e.map(float), #"p34",

            0x555555, p41e.map(float), #"p41",
            0x555555, p42e.map(float), #"p42",
            0x555555, p43e.map(float), #"p43",
            0x555555, p44e.map(float), #"p44",

            0x555555, p51e.map(float), #"p51",
            0x555555, p52e.map(float), #"p52",
            0x555555, p53e.map(float), #"p53",
            0x555555, p54e.map(float), #"p54",

            0x555555, p61e.map(float), #"p61",
            0x555555, p62e.map(float), #"p62",
            0x555555, p63e.map(float), #"p63",
            0x555555, p64e.map(float), #"p64",
    
            # Ring-Cables
            0x555555, [cr1[0][0].map(float), cr1[0][1].map(float)], #"cr11",
            0x555555, [cr1[1][0].map(float), cr1[1][1].map(float)], #"cr12",
            0x555555, [cr1[2][0].map(float), cr1[2][1].map(float)], #"cr13",
            0x555555, [cr2[0][0].map(float), cr2[0][1].map(float)], #"cr21",
            0x555555, [cr2[1][0].map(float), cr2[1][1].map(float)], #"cr22",
            0x555555, [cr2[1][0].map(float), cr2[1][1].map(float)], #"cr23",
            0x555555, [cr3[0][0].map(float), cr3[0][1].map(float)], #"cr31",
            0x555555, [cr3[1][0].map(float), cr3[1][1].map(float)], #"cr32",
            0x555555, [cr3[2][0].map(float), cr3[2][1].map(float)], #"cr33",
            0x555555, [cr4[0][0].map(float), cr4[0][1].map(float)], #"cr41",
            0x555555, [cr4[1][0].map(float), cr4[1][1].map(float)], #"cr42",
            0x555555, [cr4[2][0].map(float), cr4[2][1].map(float)], #"cr43",
            0x555555, [cr5[0][0].map(float), cr5[0][1].map(float)], #"cr51",
            0x555555, [cr5[1][0].map(float), cr5[1][1].map(float)], #"cr52",
            0x555555, [cr5[2][0].map(float), cr5[2][1].map(float)], #"cr53",
            0x555555, [cr6[0][0].map(float), cr6[0][1].map(float)], #"cr61",
            0x555555, [cr6[1][0].map(float), cr6[1][1].map(float)], #"cr62",
            0x555555, [cr6[2][0].map(float), cr6[2][1].map(float)], #"cr63",
            0x555555, [cr7[0][0].map(float), cr7[0][1].map(float)], #"cr71",
            0x555555, [cr7[1][0].map(float), cr7[1][1].map(float)], #"cr72",
            0x555555, [cr7[2][0].map(float), cr7[2][1].map(float)], #"cr73",
            0x555555, [cr8[0][0].map(float), cr8[0][1].map(float)], #"cr81",
            0x555555, [cr8[1][0].map(float), cr8[1][1].map(float)], #"cr82",
            0x555555, [cr8[2][0].map(float), cr8[2][1].map(float)], #"cr83",

            # Edge-Cables
            0x555555, ce1, #"ce1",
            0x555555, ce2, #"ce2",
            0x555555, ce3, #"ce3",
            0x555555, ce4, #"ce4",
            0x555555, ce5, #"ce5",
            0x555555, ce6, #"ce6",
            0x555555, ce7, #"ce7",
            0x555555, ce8, #"ce8",
            0x555555, ce9, #"ce9",
            0x555555, ce10, #"ce10",
            0x555555, ce11, #"ce11",
            0x555555, ce12, #"ce12",

            # Midpoints 
            0x0000aa, m1e.map(float), "center_cable_ring",
            0x0000aa, mp1e.map(float), "center plate",
           
            # Lines of Plates
            0x555555, pl1[1], pl1[2], pl1[3],pl1[4],
            0x555555, pl2[1], pl2[2], pl2[3],pl2[4],
            0x555555, pl3[1], pl3[2], pl3[3],pl3[4],
            0x555555, pl4[1], pl4[2], pl4[3],pl4[4],
            0x555555, pl5[1], pl5[2], pl5[3],pl5[4],
            0x555555, pl6[1], pl6[2], pl6[3],pl6[4],

            # forces at p11
            0xaa0000, [p11e.map(float), p1f1.map(float)],"f1", 
            0xaa0000, [p11e.map(float), p1f2.map(float)],"f2",
            '<G opacity="0.8" stroke-dasharray="0.02,0.007">',
            0xaa0000, [p1f1.map(float), p1f12.map(float)],
            0xaa0000, [p1f2.map(float), p1f12.map(float)],
            0xaa0000, [m1e.map(float), p1f12.map(float)],
            0xaa0000, [mp1e.map(float), p1f3.map(float)],
            0xaa0000, [mp1e.map(float), p1f12.map(float)],
            '</G>',
            0xaa0000, [p11e.map(float), p1f12.map(float)],"f12",
            0xaa0000, [p11e.map(float), p1f3.map(float)],"f3",
            0x009900, [p11e.map(float), p1f4.map(float)],"f4",

            # Origin   
            #0xff0000, p00.map(float), "center hexahedron",

        ]
    # Some setup for camera 
    # Rotors use half-angles:
    # Angles in radians: 30° = pi/6, 15° = pi/12
    # Rotors use half-angles:
    hx = 0
    hy = 0  
    hz = 0 
    Rx = cos(hx) - sin(hx) * e23  # Rotate around X (YZ plane)
    Ry = cos(hy) - sin(hy) * e13  # Rotate around Y (ZX plane)
    Rz = cos(hz) - sin(hz) * e12  # Rotate around Z (XY plane)
    
    # Multiply rotors to get the final camera perspective
    camera_motor = Rz * Ry * Rx


    return alg.graph(
        graph_func,
        lineWidth=1,
        grid=False,
        labels=False,
        pointRadius=0.5,
        gl=0,
        fontSize=0.6,
        scale=1.1,
        height='1000px'
    )
```
